### 一、得到所有网址链接

定义函数allurl()封装得到的网页链接

```python
def allurl():
    html = []
    for i in range(10):
        url = 'https://book.douban.com/top250?start=%d'%(i*25)
        html.append(url)
    return html
```

### 二、获取网页源代码内容

```python
import requests
url = 'https://book.douban.com/top250?start='
r = requests.get(url)
print(r.status_code)
```

```python
print(r.reason)
```

定义函数gethtmltext()封装获取网页源代码内容,以第一页为例：

```python
def gethtmltext(url):
    try:
        r = requests.get(url,timeout = 30)
        return r.text
    except:
        print("获取网页失败")
```

```python
#将以上封装成函数
import re
import pandas as pd
from pandas import DataFrame
def getbookinfo(url):
    html = gethtmltext(url)
    title = re.findall(r'title=".*?"\n',html)
    rate = re.findall(r' <span class="rating_nums">[\d\.]*</span>',html)
    vote = re.findall(r'[\d]+人评价',html)
    for i in range(len(title)):
        title[i] =  re.search(r'\"(.+)\"', title[i]).group(1)
        rate[i] = re.sub(r'<.*?>','',rate[i])
        vote[i] = re.sub(r'人评价','',vote[i]) 
    data = {'title': title, 'rate': rate, 'vote': vote}  
    frame = DataFrame(data, columns=['title', 'rate', 'vote'])
    return frame
```

### 四、将三个函数整合成一个函数

```python
import pandas as pd
from pandas import DataFrame
df = {'title':[],'rate':[],'vote':[]}
all = DataFrame(df)
```

```python
for url in allurl():
    gethtmltext(url)
    frame = getbookinfo(url)
    all = pd.concat([all,frame],ignore_index=True)
```
